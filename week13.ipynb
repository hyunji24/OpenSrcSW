```python
# 실습1: 생일 기온 시각화

import matplotlib.pyplot as plt
import csv

f=open('seoul10.csv')
data=csv.reader(f)
next(data)
high=[]
low=[]
for row in data:
    if row[-1]!='':
        if row[0].split('-')[1]=='01' and row[0].split('-')[2]=='22':
            high.append(float(row[-1]))
    if row[-2]!='':
        if row[0].split('-')[1]=='01' and row[0].split('-')[2]=='22':
            low.append(float(row[-2]))
    

plt.plot(high, color='violet')
plt.plot(low, color='lightskyblue')
plt.show()
```

```python
# 실습2: 최고 기온 시각화

import matplotlib.pyplot as plt
import csv

f=open('seoul10.csv')
data=csv.reader(f)
next(data)
jan=[]
aug=[]

for row in data:
    month=row[0].split('-')[1]
    if row[-1]!='':
        if month=='01':
            jan.append(float(row[-1]))
        if month=='08':
            aug.append(float(row[-1]))

plt.hist(jan,bins=100,color='b',label='Jan')
plt.hist(aug,bins=100,color='r',label='Aug')
plt.legend(loc='upper right')
plt.show()
```
