```python
class Calc:
    #subtraction
    def sub(self,a,b):
        result=a-b
        print('{0} - {1} = {2}'.format(a,b,result))
        
    #multiplication
    def mul(self,a,b):
        result=a*b
        print('{0} * {1} = {2}'.format(a,b,result))
        
    #division
    def div(self,a,b):
        result=a/b
        print('{0} / {1} = {2}'.format(a,b,result))
```

```python
class Stack:
    def __init__(self):
        self.data = []

    def size(self):
        return len(self.data)

    def isEmpty(self):
        return self.size() == 0

    def push(self, item):
        self.data.append(item)

    def pop(self):
        return self.data.pop()

    def peek(self):
        return self.data[-1]


def splitTokens(exprStr):
    tokens = []
    val = 0                 
    valProcessing = False
    
    for c in exprStr:
        if c == ' ':
            continue

        if c in '0123456789' :
            val = val * 10 + int(c)
            valProcessing = True
        
        else:
            if valProcessing:
                tokens.append(val)
                val = 0
            
            valProcessing = False
            tokens.append(c)
        
    if valProcessing:
        tokens.append(val)
    return tokens


def infixToPostfix(tokenList):
    prec = {
        '*' : 3,
        '/' : 3,
        '+' : 2,
        '-' : 2,
        '(' : 1
    }
    
    op = Stack()
    postfixList = []
    
    for token in tokenList:
        if type(token) is int:
            postfixList.append(token)
            
        elif token == '(':
            op.push(token)
            
        elif token == ')':
            while op.peek() != '(':
                postfixList.append(op.pop())
            op.pop()
        
        else:
            if op.isEmpty():
                op.push(token)
                
            else:
                
                while op.size() > 0:
                    if prec[op.peek()] >= prec[token]:
                        postfixList.append(op.pop())
                    else:
                        break
                
                op.push(token)
    
    while not op.isEmpty():
        postfixList.append(op.pop())
    
    return postfixList



def postfixEval(tokenList):
    num = Stack()
    
    for token in tokenList:
        if type(token) is int:
            num.push(token)
            
        elif token == '+':
            n1 = num.pop()
            n2 = num.pop()
            num.push(n2+n1)
        
        elif token == '-':
            n1 = num.pop()
            n2 = num.pop()
            num.push(n2-n1)
            
        elif token == '*':
            n1 = num.pop()
            n2 = num.pop()
            num.push(n2*n1)
            
        elif token == '/':
            n1 = num.pop()
            n2 = num.pop()
            num.push(int(n2/n1))
        
    return num.pop()

userinput = input()
tokens = splitTokens(userinput)
postfix = infixToPostfix(tokens)
val = postfixEval(postfix)


for t in tokens:
    print(t,end='')
print("=",end='')
print(val)
```
