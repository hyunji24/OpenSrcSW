```python
import numpy as np
import csv
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')

df=pd.read_csv('people.csv',engine='python',encoding='CP949',header=0,index_col=0,thousands=',')
df=df.drop(['2020년12월_계_총인구수','2020년12월_계_연령구간인구수'],axis=1)
df=df.drop(index='서울특별시  (1100000000)')


name='광진구'
a=df.index.str.contains(name)
df2=df[a]


x=df.sub(df2.iloc[0],axis=1)
y=np.power(x,2)
y['sum']=np.sum(y,axis=1)
z=y.sort_values('sum')
del z['sum']
#i=z.sort_values().index[0:2]
result=z.iloc[[0,1,-1]]






plt.style.use('ggplot')
result.T.plot()
plt.title('광진구 지역과 가장 비슷한/다른 인구 구조를 가진 지역')
plt.show()
print(result.std(axis=1))

```
